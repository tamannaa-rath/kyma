```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/features.csv")
df.head()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(
    data=df,
    x="RMS",
    y="Kurtosis",
    hue="Label",
    alpha=0.7
)

plt.title("RMS vs Kurtosis")
plt.xlabel("RMS (Overall Energy)")
plt.ylabel("Kurtosis (Spikiness)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(
    data=df,
    x="RMS",
    y="CrestFactor",
    hue="Label",
    alpha=0.7
)

plt.title("RMS vs Crest Factor")
plt.xlabel("RMS")
plt.ylabel("Crest Factor")
plt.grid(True)
plt.show()
```
